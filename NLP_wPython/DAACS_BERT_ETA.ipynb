## Reading Essay text from a local .txt file

<br>

```python
from nltk.corpus import PlaintextCorpusReader
corpus_root = '/home/bonzilla/Documents/DAACS_BERT/DAACS-Writing/data/WGU-Essays'
wordlists = PlaintextCorpusReader( corpus_root, '.*' )
wordlists.fileids()
```

```python
wordlists.words( 'Essay0001.txt' )
```

```python
file_path = corpus_root + '/Essay0001.txt'
file = open( file_path )
Essay0001_raw = file.read( )
```

```python
Essay0001_raw
```

```python
file = open( file_path, newline= '\n' )
for line in file:
    print( line.strip() )
```

```python
tokens = nltk.word_tokenize( Essay0001_raw )
text = nltk.Text( tokens )
type( text )
```

```python
print( text.collocation_list() )
print( text.concordance( 'learning' ) )
```
